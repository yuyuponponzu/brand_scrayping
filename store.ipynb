```python
import requests
import time
import urllib.parse
import urllib.request
import json
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
# セレクトショップ取得
driver = webdriver.Chrome()
driver.get('https://www.fashion-press.net/shops/select')
# 複数指定時はこの書き方
dri = driver.find_elements_by_css_selector(".columns.is-mobile")
url_prefectures = {}

# 県ごとのリンクを取得
# 県名がなぜか取れないので遷移先で拾う
for i in dri:
    print(i.get_attribute("href"))
    #tex = i.find_element_by_tag_name('a').text
    _ = []
    for j in i.find_elements_by_tag_name('div'):
        for k in i.find_elements_by_tag_name('a'):
            print(k.get_attribute("href"))
            _.append(k.get_attribute("href"))
    url_prefectures[i.find_element_by_tag_name('a').get_attribute("href")] = _
```

```python
# 各県 -> 各町 ごとに Store詳細のURLとカテゴリ、対象の性別を取得する
prefectures = {}
driver = webdriver.Chrome()
for i in url_prefectures:
    print(i)
    _ = {}
    # 各町のurlを探索
    for j in url_prefectures[i]:
        driver.get(j)
        _[driver.find_element_by_xpath("/html/body/div[5]/div/div[1]/div[1]/div[3]/div[1]/button").text.split()[2]]={}
        # 念のために探索もとのURLを入れる
        _[driver.find_element_by_xpath("/html/body/div[5]/div/div[1]/div[1]/div[3]/div[1]/button").text.split()[2]]["base_url"] = j
        # その町のStoreのURLを格納
        stores = {}
        for k in driver.find_element_by_class_name("map_wrapper").find_elements_by_class_name("other_map"):
            print(k.find_element_by_class_name("other_map_name").text)
            stores[k.find_element_by_class_name("other_map_name").text] = {}
            # 次に遷移すべきページURL
            stores[k.find_element_by_class_name("other_map_name").text]["store_url"] = k.find_element_by_class_name("other_map_name").get_attribute("href")
            try:
                # ジャンル情報
                for l in k.find_element_by_class_name("genre_list").find_elements_by_tag_name("span"):
                    # 対象性別を取得
                    if l.text == "メンズ" or l.text == "ウィメンズ":
                        stores[k.find_element_by_class_name("other_map_name").text]["sex"] = [*stores[k.find_element_by_class_name("other_map_name").text]["sex"], l.text] if "sex" in stores[k.find_element_by_class_name("other_map_name").text] else [l.text]
                    # ジャンルを取得
                    else : 
                        stores[k.find_element_by_class_name("other_map_name").text]["genre"] = [*stores[k.find_element_by_class_name("other_map_name").text]["genre"], l.text] if "genre" in stores[k.find_element_by_class_name("other_map_name").text] else [l.text]
            except:
                print("ジャンル取得に失敗しました")
                print(k.find_element_by_class_name("other_map_name").text)

        _[driver.find_element_by_xpath("/html/body/div[5]/div/div[1]/div[1]/div[3]/div[1]/button").text.split()[2]]["stores"] = stores
    # 県の下に町ごとのデータを格納する。
    prefectures[driver.find_element_by_xpath("/html/body/div[5]/div/div[1]/div[1]/div[3]/div[1]/button").text.split()[0]] =  _
```

```python
# Storeごとの情報を探索し、追加する
# 各県
driver = webdriver.Chrome()
# ブランドリストを作成
brands = set()
for i in prefectures:
    # 各町
    for j in prefectures[i]:
        # Store詳細のURL探索
        for k in prefectures[i][j]["stores"]:
            driver.get(prefectures[i][j]["stores"][k]["store_url"])
            dri = driver.find_element_by_css_selector(".columns.is-multiline.is-mobile")
            prefectures[i][j]["stores"][k]["map_url"] = dri.find_element_by_tag_name("iframe").get_attribute("href")
            
            # Store詳細情報取得
            store_detail = {}
            # 住所情報取得
            store_detail["address"]= dri.find_element_by_css_selector(".column.is-three-quarters.column0").text
            # 各項目を取得
            clms = []
            vals = []
            refs = []
            for ___ in dri.find_elements_by_css_selector(".column.is-one-quarter"):
                clms.append(___.text)
            for ___ in dri.find_elements_by_css_selector(".column.is-three-quarters"):
                vals.append(___.text)
                try:
                    refs.append(___.get_attribute("href"))
                except:
                    refs.append("")
            for clm, val, ref in zip(clms, vals, refs):
                store_detail[clm] = {}
                store_detail[clm]["val"] = val
                if ref != "":
                    store_detail[clm]["href"] = ref

            # 取り扱いブランド取得（一回ずつアクセスするんではなくて、後でまとめてブランド情報は取得する。）
            handle_brands = []
            try:
                for l in driver.find_element_by_id("b_tags").find_element_by_class_name("fp_tags").find_elements_by_tag_name("a"):
                    try:
                        handle_brands.append(l.get_attribute("href"))
                        brands.add(l.get_attribute("href"))
                    except:
                        print("ブランド遷移先の情報が取れなかった")
            except:
                print("取り扱いブランド情報がない")
            store_detail["取り扱いブランド"] = handle_brands
            prefectures[i][j]["stores"][k]["detail"] = store_detail
```

```python
# jsonで保存
import json
file1 = open('prefecture_before.json','w')
trp = json.dumps(prefectures)
json.dump(trp,file1,indent=4)
```

```python
# ブランド情報取得
# ブランド名がなぜか正規の表現ではないので、遷移した場所で正規名を取得する
import time
driver = webdriver.Chrome()
brands_names = {}
for i in brands:
    driver.get(i)
    time.sleep(1)
    brands_names[i] = {}
    brands_names[i]["brand_name"] = driver.find_element_by_css_selector(".fp_h1.h_with_border.mobile_margin").text.split(driver.find_element_by_css_selector(".fp_h1.h_with_border.mobile_margin").find_element_by_tag_name("span").text)[0]
    brands_names[i]["brand_sub_name"] = driver.find_element_by_css_selector(".fp_h1.h_with_border.mobile_margin").find_element_by_tag_name("span").text
```

```python
# jsonで保存
import json
file1 = open('brand_names.json','w')
trb = json.dumps(brands_names)
json.dump(trb,file1,indent=4)
```

```python
import pickle
with open("brand_names.pkl", "wb") as f:
    pickle.dump(brands_names, f)
```

```python
# ブランド情報を置き換え
for i in prefectures:
    # 各町
    for j in prefectures[i]:
        for k in prefectures[i][j]["stores"]:
            _ = []
            __ = []
            for l in prefectures[i][j]["stores"][k]["detail"]["取り扱いブランド"]:
                _.append(brands_names[l]["brand_name"])
                __.append(brands_names[l]["brand_sub_name"])
            prefectures[i][j]["stores"][k]["detail"]["取り扱いブランド名"] = _
            prefectures[i][j]["stores"][k]["detail"]["取り扱いブランドサブ名"] = __
```
