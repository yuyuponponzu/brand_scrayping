```python
import requests
import time
import urllib.parse
import urllib.request
import json
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
# ブランドリスト取得
driver.quit()
driver = webdriver.Chrome()
driver.get('https://www.fashionsnap.com/collection/')
driver.find_element_by_link_text("ブランド").click()
brand_list = []
for i in driver.find_elements_by_class_name("letter-group"):
    for j in i.find_elements_by_tag_name('a'):
        brand_list.append(j.text)#.lower())
```

```python
# WEARでスクレイピング
# driver = webdriver.Chrome()
# driver.get('https://wear.jp/brand/')
# driver.find_element_by_id("search_container").send_keys("lad musician")
```

```python
# WEARでbrand_idを取得する
brand_ids = {}
for i in brand_list:
    SEARCH_BRAND = "https://wear.jp/sp/common/json/brand_search.json?brand_name="
    enc_name = urllib.parse.quote(i)
    url = SEARCH_BRAND + enc_name
    print(url)
    req = urllib.request.Request(url)
    with urllib.request.urlopen(req) as res:
        body = json.load(res)
        # 取得失敗およびlistサイズが0のときは存在しないブランド。無視する。
        if body[0]["result"] == "success" and len(body[0]["list"]) >= 1:
            if i in brand_ids:
                brand_ids[i].add(body[0]["list"][0]["id"])
            else :
                brand_ids[i] = set()
                brand_ids[i].add(body[0]["list"][0]["id"])
```

```python
# WEARでbrand_idからコーディネート一覧に飛びコーディネート詳細のURLリストを取得
driver.quit()
driver = webdriver.Chrome()
coord_list = {}
# WEARに存在しない場合はこれ以降除外される
for i in brand_ids.keys():
    for l in brand_ids[i]:
        coord_url = "https://wear.jp/coordinate/?brand_id=" + l
        driver.get(coord_url)
        for j in driver.find_elements_by_class_name("like_mark"):
            if i in coord_list:
                coord_list[i].add(j.find_element_by_tag_name("a").get_attribute("href"))
            else :
                coord_list[i] = set()
                coord_list[i].add(j.find_element_by_tag_name("a").get_attribute("href"))
```

```python
# 他に着用されているブランドを取得する。
driver = webdriver.Chrome()
related_brands = {}
related_brands_url = {}
error_list =[]
for i in coord_list.keys():
    for l in coord_list[i]:
        driver.get(l)
        for j in driver.find_elements_by_class_name("list, clearfix"):
            # 既に追加されてる場合とされていない場合で、初期化処理の実行有無を確認
            try:
                if i in related_brands and len(related_brands[i]) >= 1:
                    # URL設定されてないやつは基本的にゴミだから捨てる
                    related_brands_url[i].add(j.find_element_by_class_name("brand").find_element_by_tag_name("a").get_attribute("href"))
                    related_brands[i].add(j.find_element_by_class_name("brand").text)
                else:
                    related_brands[i] = set ()
                    related_brands_url[i] = set ()
                    related_brands_url[i].add(j.find_element_by_class_name("brand").find_element_by_tag_name("a").get_attribute("href"))
                    related_brands[i].add(j.find_element_by_class_name("brand").text)
            except:
                print("失敗しました")
                try:
                    print(j.find_element_by_class_name("brand").text)
                    error_list.append([i,j,j.find_element_by_class_name("brand").text])
                except:
                    print("textが設定されていない")
                    error_list.append([i,j,None])
```

```python
# jsonで保存
import json
file1 = open('brand.json','w')
json.dump(related_brands,file1,indent=4)
```
